### Retriving information from Twitter, analyzing the information and storing in a pickle file

```python
import tweepy
```

```python
# Let's hit up twitter!
from twitter_keys import consumer_key, consumer_secret, access_token, access_secret
import pickle
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token,access_secret)

api = tweepy.API(auth)

rgv_tweets = api.user_timeline('RGVzoomin')

with open('rgv_tweets.pkl','wb') as f:
    pickle.dump(rgv_tweets,f)
    
cbn_tweets = api.user_timeline('ncbn')

with open('cbn_tweets.pkl','wb') as f:
    pickle.dump(cbn_tweets,f)
    
pk_tweets = api.user_timeline('PawanKalyan')

with open('pk_tweets.pkl','wb') as f:
    pickle.dump(pk_tweets,f)
    

    
```

```python
def lexical_diversity(text):
    tokens = text.lower().split()
    return len(set(tokens))/len(tokens)

print(rgv_tweets[0].text)
print('lexical diversity for RGV tweet: ', lexical_diversity(rgv_tweets[0].text))
print(cbn_tweets[0].text)
print('lexical diversity for CBN tweet: ', lexical_diversity(cbn_tweets[0].text))
print(pk_tweets[0].text)
print('lexical diversity for PK tweet: ', lexical_diversity(pk_tweets[0].text))

```

# Sentiment Analysis

```python
import pickle 
from nltk_helpers import get_sentiments 
    
print(get_sentiments(rgv_tweets[0].text))
print(get_sentiments(cbn_tweets[0].text))
print(get_sentiments(pk_tweets[0].text))

# visualization part
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
# build our figure
fig = plt.figure()
# build the axes
ax = fig.add_subplot(111, projection='3d')

import pickle
from nltk_helpers import split_sentiments

rgvX, rgvY, rgvZ = split_sentiments([get_sentiments(tweet.text) for tweet in rgv_tweets])
ax.scatter(rgvX, rgvY, rgvZ, color = 'r', marker = 'd')

cbnX, cbnY, cbnZ = split_sentiments([get_sentiments(tweet.text) for tweet in cbn_tweets])
ax.scatter(cbnX, cbnY, cbnZ, color = 'g', marker = 'o')

pkX, pkY, pkZ = split_sentiments([get_sentiments(tweet.text) for tweet in pk_tweets])
ax.scatter(pkX, pkY, pkZ, color = 'b', marker = 's')


ax.set_xlabel('Negative')
ax.set_ylabel('Neutral')
ax.set_zlabel('Positive')

plt.show()
```

## Two Dimensional

```python
plt.scatter(rgvX, rgvY, color = 'r', marker = 'd', label = 'Ram Gopal Varma')
plt.scatter(cbnX, cbnY, color = 'g', marker = 'o', label = 'Chandra Babu Naidu')
plt.scatter(pkX, pkY, color = 'b', marker = 'p', label = 'Pawan Kalyan')
plt.xlabel('Negative')
plt.ylabel('Positive')
plt.title('Positive vs Negative Sentiments')
plt.legend(loc = 'upper right', shadow = True)
plt.show()
```

# Devices Used 

```python
print('--------------------------------RGV-------------------------------')
rgv_iphone=0
rgv_android=0 
rgv_desktop = 0
for i in rgv_tweets:
    if i.source=='Twitter for iPhone':
        rgv_iphone += 1
    elif i.source=='Twitter for Android':
        rgv_android += 1
    elif i.source=='Twitter for Desktop':
        rgv_desktop += 1
print(rgv_iphone, rgv_android, rgv_desktop)

print('--------------------------------CBN-------------------------------')
cbn_iphone=0
cbn_android=0
cbn_desktop = 0
for i in cbn_tweets:
    if i.source=='Twitter for iPhone':
        cbn_iphone += 1
    elif i.source=='Twitter for Android':
        cbn_android += 1
    elif i.source=='Twitter for Desktop':
        cbn_desktop += 1
print(cbn_iphone, cbn_android, cbn_desktop)
    
print('--------------------------------PK-------------------------------')
pk_iphone=0
pk_android=0
pk_desktop = 0
for i in pk_tweets:
    if i.source=='Twitter for iPhone':
        pk_iphone += 1
    elif i.source=='Twitter for Android':
        pk_android += 1
    elif i.source=='Twitter for Desktop':
        pk_desktop += 1
print(pk_iphone, pk_android, pk_desktop)
```

# Visualization for Devices used

```python
import numpy as np
import matplotlib.pyplot as plt

N = 3
ind = np.arange(N)  # the x locations for the groups
width = 0.27       # the width of the bars

fig = plt.figure()
ax = fig.add_subplot(111)
iphone = [rgv_iphone,cbn_iphone,pk_iphone]
rects1 = ax.bar(ind, iphone, width, color='b')
android = [rgv_android,cbn_android,pk_android]
rects2 = ax.bar(ind+width, android, width, color='r')
web = [rgv_desktop, cbn_desktop, pk_desktop] 
rects3 = ax.bar(ind+width*2, web, width, color='g')

ax.set_ylabel('No of users')
ax.set_xlabel('Accounts')
ax.set_xticks(ind+width)
ax.set_xticklabels( ('RGV', 'CBN', 'PK') )
ax.legend( (rects1[0], rects2[0], rects3[0]), ('IPhone', 'Android', 'Desktop') )

def autolabel(rects):
    for rect in rects:
        h = rect.get_height()
        ax.text(rect.get_x()+rect.get_width()/2., 1.05*h, '%d'%int(h),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

plt.title('Devices used')
plt.show()
```
